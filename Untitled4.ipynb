```python
import pandas as pd
df = pd.read_csv("Cricket_data.csv")
df.drop(["Unnamed: 0","Unnamed: 13"],axis=1,inplace=True)
def stats(df,check):
  d = {"Players":df["Player"],check:df[check]}
  finaldf = pd.DataFrame(d)
  print(finaldf.columns)
  finaldf = finaldf.sort_values(check,ascending=False)
  finaldf = finaldf.head(10)
  li = ["He is Number one"]
  j = 0
  for i in range(1,10):
    if check=="50":
      li.append("{0} needs {1} fifties to acheive {2} Position".format(finaldf.iloc[i,0],finaldf.iloc[j,1]-finaldf.iloc[i,1],i))
    elif check=="100":
      li.append("{0} needs {1} hundreds to acheive {2} Position".format(finaldf.iloc[i,0],finaldf.iloc[j,1]-finaldf.iloc[i,1],i))
    else:
      li.append("{0} needs {1} runs to acheive {2} Position".format(finaldf.iloc[i,0],finaldf.iloc[j,1]-finaldf.iloc[i,1],i))
    j+=1
  finaldf["Needs to Acheive"] = li
  print(finaldf.drop(check,axis=1))
stats(df,"50")
stats(df,"100")
stats(df,"Runs")
```

```python
import pandas as pd
df = pd.read_csv("64865.csv")
print(df.columns)
def batsmanScorecard(df,phase):
  batsmans = set(df[df["Country"]==phase]["Batsman"])
  batscorecard = pd.DataFrame(columns=["Batsman","Runs","4","6"])
  d = {"Batsman":"","Runs":0,"4":0,"6":0}
  print(phase+" Batting Scorecard")
  for i in batsmans:
    d["Batsman"]=i
    d["Runs"]=sum(df[df["Batsman"]==i]["Runs"])
    d["4"]=len(df[(df["Batsman"]==i) & (df["Runs"]==4)])
    d["6"]=len(df[(df["Batsman"]==i) & (df["Runs"]==6)])
    batscorecard = batscorecard.append(d,ignore_index=True)
  print(batscorecard)
batsmanScorecard(df,"South Africa")
batsmanScorecard(df,"West Indies")
```
